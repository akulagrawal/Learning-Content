```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
mat = np.matrix
mat = np.loadtxt('data.txt', skiprows = 1)
```

```python
n = len(mat[0])
m = len(mat[:,0])
print(m)
color =  ['red'] * m
for i in range(0, m):
    if mat[i][0] == 2:
        color[i] = 'blue'
for i in range(1, n):
    for j in range(i+1, n):
        plt.xlabel('Feature '+str(i))
        plt.ylabel('Feature '+str(j))
        plt.scatter(mat[:,i], mat[:,j], c=color)
        plt.show()
```

```python
print('From Graph, it is visible that required features are feature 1 and feature 2')
```

```python
pca = PCA(n_components = 2)
sol = pca.fit(np.transpose(mat))
solmat = sol.components_
plt.scatter(solmat[0], solmat[1], c=color)
plt.show()
plt.scatter(mat[:,1], mat[:,2], c=color)
plt.show()
```
